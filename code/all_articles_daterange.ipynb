```python
        import json
        import urllib.request
        from urllib.request import Request, urlopen
        import sys
        sys.setdefaultencoding=('utf-8')
    
        url="https://www.bhaskar.com/sports/"
        webpage={
        "count":"",
        "source":"DAINIK BHASKAR",
        "id":"db",
        "url":url,
        "articles":[]
        }
        count=0
        #decrease the value of j as you increase the no. of days as it represents the number of pages we are gonna search the required date
        j=13

        links=[]
        b=[]
        while j<16:
        
            req = Request(url+"{}".format(j), headers={'User-Agent': 'Chrome/5.0'})
            open_url=urllib.request.urlopen(req)
            from bs4 import BeautifulSoup
            content=BeautifulSoup(open_url,"html.parser")
            news_links=content.find_all("a",attrs={"class":"list_text"})


            for link in news_links:
                links.append(link['href'])

            for link in links:
                if link[1:7]=="sports":
                    b.append("https://www.bhaskar.com"+link)
            j+=1
```

```python
#starting date and ending date is given.Month is given directly in condition
#this code is fetching all news articles between 31 oct and 1 november both inclusive

start,end=31,1
for i in b:
    req = Request(i, headers={'User-Agent': 'Chrome/5.0'})
    open_url=urllib.request.urlopen(req)
    from bs4 import BeautifulSoup
    main_content=BeautifulSoup(open_url,"html.parser")
    date_div=main_content.find('div',class_='db_storytime')
    string=date_div.text
    l=string[15:21].split()
    l[1]=int(l[1])
    if (l[0]=='Nov' and l[1]<=end) or (l[0]=='Oct' and l[1]>=start):
        r=Request(i, headers={'User-Agent': 'Chrome/5.0'})
        open_link=urllib.request.urlopen(r)
        from bs4 import BeautifulSoup
        main_content=BeautifulSoup(open_link,"html.parser")
        page={"ID:":"db",
                "URL OF PAGE:" :link,
                "SOURCE | DATE & TIME:" :"",
                "TITLE:" :"",
                "IMAGE URLs:" :[],
                "IMAGE CAPTIONs:" :[],
                "TEXT:":[],
                "CATEGORY:" :[]}
        source_date_time=main_content.find('div',attrs={"class":"db_storytime"})
        page['SOURCE | DATE & TIME:']= source_date_time.get_text()
        title=main_content.find("title")
        page['TITLE:']=title.get_text()

        image_url=main_content.find_all("img")
        for i in image_url:
            page["IMAGE URLs:"].append(i.get('src'))


        image_captions=main_content.find_all("span",attrs={"class":"db_slug"})
        for i in image_captions:
            page['IMAGE CAPTIONs:'].append(i.get_text())

        text=main_content.find_all('div',attrs={"class":"db_storycontent"})
        for i in range(0,len(text)):
            page['TEXT:'].append(text[i].get_text())

        category=main_content.find_all('div',attrs={"class":"db_artbredcram"})
        for i in range(0,len(category)):
            page['CATEGORY:'].append(category[i].get_text())


        webpage['articles'].append(page)
        count=count+1
        webpage['count']=count
with open("all_articles_daterange.json",'w') as f:
    json.dump(webpage,f,indent=2,ensure_ascii=False)
```
