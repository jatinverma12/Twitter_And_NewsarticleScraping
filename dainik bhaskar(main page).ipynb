```python
import json
import urllib.request
from urllib.request import Request, urlopen
website='https://www.bhaskar.com/sports/anya-khel/news/atp-challenger-sumit-nagal-enters-pre-quarter-final-tennis-news-update-01656839.html'

req = Request('https://www.bhaskar.com/sports/anya-khel/news/atp-challenger-sumit-nagal-enters-pre-quarter-final-tennis-news-update-01656839.html', headers={'User-Agent': 'Chrome/5.0'})
#webpage = urlopen(req).read()
```

```python
open_website=urllib.request.urlopen(req)
from bs4 import BeautifulSoup
main_content=BeautifulSoup(open_website,"html.parser")
print(main_content)
```

```python
f=open("main_content",'w', encoding='UTF-8')
f.write("URL OF PAGE : "+website+"\n\n\n")
f.close()
```

```python
source_date_time=main_content.find('div',attrs={"class":"db_storytime"})
f=open("main_content",'a', encoding='UTF-8')
f.write("SOURCE | DATE & TIME :" + source_date_time.get_text() + '\n\n\n' )
f.close()
```

```python
title=main_content.find("title")
```

```python
f=open("main_content",'a',encoding="UTF-8")
f.write("TITLE : " + title.get_text() +'\n\n\n')
f.close()
```

```python
image_url=main_content.find_all("img")
f=open("main_content",'a',encoding="UTF-8")
f.write("IMAGE URLs :\n")
image_urls=[]
for image in image_url:
   #3image_urls.append(image['src'])
    f.write(image['src']+'\n')
    #f.write(image_urls)
f.write('\n\n\n')
f.close()
```

```python
text=main_content.find_all('div',attrs={"class":"db_storycontent"})
f=open("main_content",'a', encoding='UTF-8')
f.write("TEXT :")
for i in range(0,len(text)):
     f.write(text[i].get_text()+'\n')
f.close()
```

```python
category=main_content.find_all('div',attrs={"class":"db_artbredcram"})
f=open("main_content",'a', encoding='UTF-8')
f.write("CATEGORY :")
for i in range(0,len(category)):
     f.write(category[i].get_text()+'\n')
f.close()
```

```python
f=open("main_content",'r',encoding="UTF-8")
content=f.read()
with open("main_content.json",'w') as f:
        json.dump(content,f)
```
